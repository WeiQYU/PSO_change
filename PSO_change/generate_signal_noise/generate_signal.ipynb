```python
import matplotlib.pyplot as plt
from scipy.io import savemat
from scipy.signal import welch
import numpy as np
```

```python

# 生成一个100个周期，振幅为1e-18

def generate_signal(A,coefs,times):
    signal = np.sin(2 * np.pi * coefs[0] * times + coefs[1])
    signal = A * signal / np.linalg.norm(signal)
    return signal

A = 1e-10
nsamples = 2 ** 14
coefs = np.array([0.05,0.5])
timeVec = np.linspace(0,2000,nsamples)
signal = generate_signal(A,coefs,timeVec)

mask = (timeVec >= 500) & (timeVec <= 750)

signal = signal * mask

savemat('signal.mat',{'signal':signal})

# print(1 if (1e-11 == 10 ** (-11)) else 0)

fig = plt.figure()
plt.plot(timeVec,signal)
# plt.scatter(timeVec,signal)
plt.show()



# f,pxx = welch(signal,nperseg=nsamples)
# fig = plt.figure()
# plt.plot(f,pxx)
# plt.yscale('log')
# plt.show()
```
